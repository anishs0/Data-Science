```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
#Boston datasets 13 numerical features and a numerical target variable
boston_dataset = datasets.load_boston()
X_full = boston_dataset.data
Y = boston_dataset.target
print (X_full.shape)
print(Y.shape)
```

```python
print (boston_dataset.DESCR)
```

```python
selector = SelectKBest(f_regression, k=1)
selector.fit(X_full, Y)
X = X_full[:, selector.get_support()]
print (X.shape)
```

```python
def plot_scatter(X,Y,R=None):
    plt.scatter(X, Y, s=32, marker='o', facecolors = "blue")
    if R is not None:
            plt.scatter(X,R, color = "red", linewidth = 0.5)
    plt.show()
plot_scatter(X,Y)
```

```python
regressor = LinearRegression(normalize=True).fit(X, Y)
plot_scatter(X, Y, regressor.predict(X))
```

```python
regressor = SVR().fit(X, Y)
plot_scatter(X, Y, regressor.predict(X))
```

```python
regressor = RandomForestRegressor().fit(X, Y)
plot_scatter(X, Y, regressor.predict(X))
```

### Pandas: Offers an off-the-shelf function to start exploring relationships and distributions between the quantitatives variables in a dataset:

```python
import pandas as pd
import numpy as np
colors = list()
palette = {0: "red", 1: "green", 2: "blue"}
for c in np.nditer(iris.target):
    colors.append(palette[int(c)])
dataframe = pd.DataFrame(iris.data,  columns=iris.feature_names)
sc = pd.plotting.scatter_matrix(dataframe, alpha=0.3, figsize=(10, 10), diagonal='hist', color=colors, marker='o', grid=True)
```
